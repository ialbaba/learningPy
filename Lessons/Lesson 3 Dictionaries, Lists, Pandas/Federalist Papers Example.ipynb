```python
## Opening and reading a file. ONLY .txt files for now

file_name = 'federalist_papers.txt'

paper = open(file_name, 'r')
text = paper.read()
text = text.lower()
text
```

```python
punc = [',',".", "!", "/", "?", '\n', ':', "(", ")", ";"]

for i in punc:
    text = text.replace(i," ")

word_list = text.split(" ")
```

```python
new_list = []
for word in word_list:
    if word != "":
        new_list.append(word)
```

```python
word_dict = {}

for word in new_list:
    if word in word_dict:
        word_dict[word] = word_dict[word] + 1
    else:
        word_dict[word] = 1

word_dict
```
